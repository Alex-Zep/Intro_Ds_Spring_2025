<span style="color:green; font-size:18px"><strong>9. Use .groupby() to compute summary statistics (e.g., mean or count) across
categories.</strong></span><br>

```python
if 'State' in df.columns:
    print("\nAverage Black or African American Alone Population by State:")
    print(df.groupby('State')['Black or African American Alone'].mean().sort_values(ascending=False))
```

<span style="color:red; font-size:16px"><em>This output shows the average population of Black or African American individuals, specifically those counted as "Black or African American Alone," across different states. The District of Columbia has the highest average at 290,772, followed by Maryland with about 76,078. States like New Jersey, Florida, and New York also have relatively high averages. This suggests that certain states, particularly on the East Coast, have significant Black or African American populations, which could be important for demographic studies, policy-making, or targeted community programs. Lower averages in some states may indicate smaller Black or African American populations relative to other racial groups.</em></span>


```python
if 'State' in df.columns:
    print("\nTotal Female Population by State:")
    print(df.groupby('State')['Female Population'].sum().sort_values(ascending=False))
```

<span style="color:red; font-size:16px"><em>This output displays the total female population by state, based on a grouped sum of the 'Female Population' column. California ranks highest with over 19.6 million women, followed by Texas with 14.8 million and Florida with 11.1 million. These figures reflect the overall population size of these states, which are among the most populous in the U.S. This information can be useful for understanding demographic distribution, planning gender-focused programs, or identifying states with large female populations for targeted policy or outreach efforts.</em></span>
